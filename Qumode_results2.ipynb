模拟不同mode数完整输出的结果

```python
# scale
Ns: int = 1
Nm: int = 2
NL: int = 5
NL_list = np.arange(7)

# BS
eta: float = np.pi * 0.45

# S2
s_r: float = 0.6
s_phi: float = 0

# PS 
delta_k: float = 0

# S
r_k: float = 1.4
phi_k: float = np.pi   # 取非零的那个

# loss
if_add_loss: bool = True
length : int = 5
loss_factor : float = 0.01
loss_amplitude : float = length * loss_factor
loss_amplitude_list = np.linspace(0, 0.15, 16)
if_loss_input : bool = False
loss_input : np.array = None

# additonal features
if_odd_even: bool = True
if_part_computing: bool = False


for loss_amplitude in tqdm(loss_amplitude_list):

    plt.figure(figsize=(12, 4), dpi=300)

    I3 = np.zeros(len(NL_list))
    for i in range(len(NL_list)):

        NL = NL_list[i]


        ################################
        if_add_loss: bool = False
        ################################

        a = CV.TripartiteMutualInformation(
            Ns=Ns, Nm=Nm, NL=NL,  # scale
            eta=eta,              # BS
            s_r=s_r, s_phi=s_phi, # S2
            delta_k=delta_k,      # PS
            r_k=r_k, phi_k=phi_k, # S
            if_add_loss=if_add_loss,    # loss
            if_odd_even=if_odd_even,     # additonal features
            if_part_computing=if_part_computing      # additonal features
        )
        a.compute()
        I3[i] = a.I3()

    tittle = r"BS:$\eta$={:.2f}$\pi$;  S2:$r$={:.2f},$\phi$={:.2f}$\pi$;  ".format(eta/np.pi, s_r, s_phi/np.pi) + \
                r"PS:$\delta_k$={:.2f}$\pi$;  S:$r$={:.2f},$\phi$={:.2f}$\pi$".format(delta_k/np.pi, r_k, phi_k/np.pi)
                
    if if_add_loss:
        tittle += r'; loss={:.2f}dB'.format(loss_amplitude)
    if if_odd_even:
        tittle += r'; odd-even'
    plt.plot(NL_list, I3, 
                label=tittle,
                marker='s', markersize=5)
    
    I3 = np.zeros(len(NL_list))
    for i in range(len(NL_list)):

        NL = NL_list[i]

        ################################
        if_add_loss: bool = True
        ################################

        a = CV.TripartiteMutualInformation(
            Ns=Ns, Nm=Nm, NL=NL,  # scale
            eta=eta,              # BS
            s_r=s_r, s_phi=s_phi, # S2
            delta_k=delta_k,      # PS
            r_k=r_k, phi_k=phi_k, # S
            if_add_loss=if_add_loss, loss_amplitude=loss_amplitude,    # loss
            if_odd_even=if_odd_even,     # additonal features
            if_part_computing=if_part_computing,      # additonal features
        )
        a.compute()
        I3[i] = a.I3()

    tittle = r"BS:$\eta$={:.2f}$\pi$;  S2:$r$={:.2f},$\phi$={:.2f}$\pi$;  ".format(eta/np.pi, s_r, s_phi/np.pi) + \
                r"PS:$\delta_k$={:.2f}$\pi$;  S:$r$={:.2f},$\phi$={:.2f}$\pi$".format(delta_k/np.pi, r_k, phi_k/np.pi)
                
    if if_add_loss:
        tittle += r';  loss={:.2f}dB'.format(loss_amplitude)
    if if_odd_even:
        tittle += r';  odd-even'
    plt.plot(NL_list, I3, 
                label=tittle,
                marker='s', markersize=5)

    ################################

    # # 加水平横线
    plt.axhline(y=0, color='gray', linestyle='--')
    plt.axvline(x=5, color='gray', linestyle='--')

    plt.xlabel(r'NL', fontsize=16)
    plt.ylabel(r'$I_3$', fontsize=16)
    plt.legend()
    plt.title(r'Ns={}, Nm={}, loss for each component =${}\mu m$*{:.3f}dB/$\mu m$={:.3f}dB'.format(Ns, Nm, length, loss_amplitude/length, loss_amplitude), fontsize=16)
    savename = r'Ns={}-Nm={}-Ne={}'.format(Ns,Nm,NL) + \
                r"BS--eta={:.2f}pi-S2--r={:.2f}-phi={:.2f}pi-".format(eta/np.pi, s_r, s_phi/np.pi) + \
                r"PS--delta_k={:.2f}pi-S--r={:.2f}-phi={:.2f}pi".format(delta_k/np.pi, r_k, phi_k/np.pi) + \
                r'-loss={:.2f}dB'.format(loss_amplitude)   
    plt.savefig(r'pic-NL-loss/'+savename+r'.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

    from IPython.display import clear_output
    # 清除输出
    clear_output(wait=True)

print('All are finished!')

```

```python
# scale
Ns: int = 1
Nm: int = 2
NL: int = 5

# BS
eta: float = np.pi * np.random.rand()
eta_list = np.linspace(0, np.pi, 101)

# S2
s_r: float = 0.6
s_phi: float = 0

# PS 
delta_k: float = 0

# S
r_k: float = 1.4
phi_k: float = np.pi   # 取非零的那个

# loss
if_add_loss: bool = True
length : int = 5
loss_factor : float = 0.01
loss_amplitude : float = length * loss_factor
loss_amplitude_list = np.linspace(0, 0.15, 16)
if_loss_input : bool = False
loss_input : np.array = None

# additonal features
if_odd_even: bool = True
if_part_computing: bool = True
part_computing_layers: int = np.arange(4*(Nm+NL) - 5)


for eta in tqdm(eta_list):
    for loss_amplitude in loss_amplitude_list:

        I3 = np.zeros(len(part_computing_layers))

        plt.figure(figsize=(12, 4), dpi=300)
        ################################
        if_add_loss: bool = False
        ################################

        for i in range(len(part_computing_layers)):

            a = CV.TripartiteMutualInformation(
                Ns=Ns, Nm=Nm, NL=NL,  # scale
                eta=eta,              # BS
                s_r=s_r, s_phi=s_phi, # S2
                delta_k=delta_k,      # PS
                r_k=r_k, phi_k=phi_k, # S
                if_add_loss=False,    # loss
                if_odd_even=True,     # additonal features
                if_part_computing=True, part_computing_layers=part_computing_layers[i]      # additonal features
            )
            a.compute()
            I3[i] = a.I3()

        tittle = r"BS:$\eta$={:.2f}$\pi$;  S2:$r$={:.2f},$\phi$={:.2f}$\pi$;  ".format(eta/np.pi, s_r, s_phi/np.pi) + \
                    r"PS:$\delta_k$={:.2f}$\pi$;  S:$r$={:.2f},$\phi$={:.2f}$\pi$".format(delta_k/np.pi, r_k, phi_k/np.pi)
                    
        if if_add_loss:
            tittle += r'; loss={:.2f}dB'.format(loss_amplitude)
        if if_odd_even:
            tittle += r'; odd-even'
        plt.plot(part_computing_layers, I3, 
                    label=tittle,
                    marker='s', markersize=5)
        plt.legend()


        ################################
        if_add_loss: bool = True
        ################################

        for i in range(len(part_computing_layers)):

            a = CV.TripartiteMutualInformation(
                Ns=Ns, Nm=Nm, NL=NL,  # scale
                eta=eta,              # BS
                s_r=s_r, s_phi=s_phi, # S2
                delta_k=delta_k,      # PS
                r_k=r_k, phi_k=phi_k, # S
                if_add_loss=True, loss_amplitude=loss_amplitude,    # loss
                if_odd_even=True,     # additonal features
                if_part_computing=True, part_computing_layers=part_computing_layers[i]      # additonal features
            )
            a.compute()
            I3[i] = a.I3()

        tittle = r"BS:$\eta$={:.2f}$\pi$;  S2:$r$={:.2f},$\phi$={:.2f}$\pi$;  ".format(eta/np.pi, s_r, s_phi/np.pi) + \
                    r"PS:$\delta_k$={:.2f}$\pi$;  S:$r$={:.2f},$\phi$={:.2f}$\pi$".format(delta_k/np.pi, r_k, phi_k/np.pi)
                    
        if if_add_loss:
            tittle += r';  loss={:.2f}dB'.format(loss_amplitude)
        if if_odd_even:
            tittle += r';  odd-even'
        plt.plot(part_computing_layers, I3, 
                    label=tittle,
                    marker='s', markersize=5)
        plt.legend()

        ################################

        # # 加水平横线
        plt.axhline(y=0, color='gray', linestyle='--')
        # plt.axvline(x=5, color='gray', linestyle='--')

        plt.xlabel(r'layers', fontsize=16)
        plt.ylabel(r'$I_3$', fontsize=16)
        plt.legend()
        plt.title(r'Ns={}, Nm={}, Ne={}, loss for each component =${}\mu m$*{:.2f}dB/$\mu m$={:.2f}dB'.format(Ns, Nm, NL, length, loss_amplitude/length, loss_amplitude), fontsize=16)
        savename = r"BS--eta={:.2f}pi-S2--r={:.2f}-phi={:.2f}pi-".format(eta/np.pi, s_r, s_phi/np.pi) + \
                    r"PS--delta_k={:.2f}pi-S--r={:.2f}-phi={:.2f}pi".format(delta_k/np.pi, r_k, phi_k/np.pi) + \
                    r'-loss={:.2f}dB'.format(loss_amplitude)   
        plt.savefig(r'pic-bs-loss/'+savename+r'.png', 
                    dpi=300, bbox_inches='tight')
        plt.show()

        from IPython.display import clear_output
        # 清除输出
    clear_output(wait=True)

print('All are finished!')
```
